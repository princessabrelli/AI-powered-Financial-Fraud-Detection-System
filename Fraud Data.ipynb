```python
# Import necessary libraries
import pandas as pd
import numpy as np

file_path = r'C:\Users\genes\documents\transactions.csv'

df = pd.read_csv(file_path)


# Data overview
print("Initial data overview:")
print(df.info())
print(df.head())

# Step 1: Clean column names (strip spaces)
df.columns = df.columns.str.strip()

# Step 2: Handle missing values in 'TRANSACTION DETAILS', 'WITHDRAWAL AMT', 'DEPOSIT AMT', 'BALANCE AMT'
df['TRANSACTION DETAILS'].fillna('Unknown', inplace=True)

# Fill missing withdrawal/deposit amounts with 0 (no transaction)
df['WITHDRAWAL AMT'].fillna('0', inplace=True)
df['DEPOSIT AMT'].fillna('0', inplace=True)

# Step 3: Combine 'WITHDRAWAL AMT' and 'DEPOSIT AMT' into 'transaction_amount'
df['WITHDRAWAL AMT'] = df['WITHDRAWAL AMT'].str.replace(',', '').astype(float)
df['DEPOSIT AMT'] = df['DEPOSIT AMT'].str.replace(',', '').astype(float)

df['transaction_amount'] = df['DEPOSIT AMT'] - df['WITHDRAWAL AMT']

# Step 4: Convert 'BALANCE AMT' to numerical values
df['BALANCE AMT'] = df['BALANCE AMT'].str.replace(',', '').astype(float)

# Step 5: Drop irrelevant column '.'
df.drop(columns=['.'], inplace=True)

# Step 6: Display the cleaned data overview
print("\nCleaned data overview:")
print(df.info())
print(df.head())

# Step 7: Save the cleaned data to a new CSV file
cleaned_file_path = r'C:\Users\genes\documents\transactions.csv'
df.to_csv(cleaned_file_path, index=False)
```
